# 이상치

```python
def outlier(x):
    q1=x.quantile(0.25)
    q3=x.quantile(0.75)
    iqr=q3 - q1
    lb=q1 - 1.5*iqr
    ub=q3 + 1.5*iqr
    
    out=(x < lb) | (x > ub)
    
    return out
```
